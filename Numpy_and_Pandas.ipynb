# **Numpy**

```python
import numpy as np
```

1. Create a NumPy array of random numbers and normalize the data using min-max scaling.

```python
arr1 = np.random.rand(10) * 100
normalized1 = (arr1 - arr1.min()) / (arr1.max() - arr1.min())
print("Array:", np.round(arr1, 2))
print("Min-max normalized:", np.round(normalized1, 3))
```

2. Generate a 5x5 matrix with random integers and replace all occurrences of a specific value with another.

```python
mat2 = np.random.randint(0, 5, size=(5, 5))
mat2_rep = np.where(mat2 == 2, 20, mat2)
print("Original matrix:\n", mat2)
print("Replaced all 2s with 20:\n", mat2_rep)
```

3. Create two NumPy arrays and perform element-wise addition, subtraction, multiplication, and division.

```python
a3 = np.array([10, 20, 30, 40])
b3 = np.array([2, 4, 5, 8])
print("a =", a3, " b =", b3)
print("Addition:", a3 + b3, "\nSubstraction:", a3 - b3, "\nMultiplication:", a3 * b3, "\nDivision:", a3 / b3)
```

4. Solve the system of linear equations: \(3x + 4y = 7\) and \(5x + 2y = 8\) using NumPy.

```python
x = np.array([[3, 4], [5, 2]])
y = np.array([7, 8])
sol = np.linalg.solve(x, y)
print("\n4. Solution (x, y):", sol)
```

5. Implement broadcasting to perform an operation between a 3x3 matrix and a 1D array.

```python
mat5 = np.arange(9).reshape(3, 3)
arr5 = np.array([1, 2, 3])
result5 = mat5 + arr5
print("Matrix:\n", mat5, "\nArray:", arr5, "\nBroadcast sum:\n", result5)
```

6. Create a 3x3 identity matrix using NumPy.

```python
id_mat = np.eye(3)
print("3x3 identity matrix:\n", id_mat)
```

7. Perform matrix multiplication between two 2D arrays.

```python
m7a = np.array([[1, 2], [3, 4]])
m7b = np.array([[5, 6], [7, 8]])
matmul7 = m7a @ m7b
print("\n7. A @ B:\n", matmul7)
```

8. Calculate the dot product and cross product of two vectors using NumPy.

```python
v8a = np.array([1, 2, 3])
v8b = np.array([4, 5, 6])
dot8 = np.dot(v8a, v8b)
cross8 = np.cross(v8a, v8b)
print("Dot product:", dot8, " Cross product:", cross8)
```

9. Generate a NumPy array of 20 random integers and find the unique elements in the array.

```python
arr9 = np.random.randint(0, 10, size=20)
unique9 = np.unique(arr9)
print("Array:", arr9)
print("\nUnique elements:", unique9)
```

10. Implement a function using NumPy that returns the inverse of a given matrix.

```python
def matrix_inverse(m):
    return np.linalg.inv(m)

mat10 = np.array([[4, 7], [2, 6]])
inv10 = matrix_inverse(mat10)
print("Matrix:\n", mat10, "\nInverse:\n", inv10)
```

26. Perform an exploratory data analysis on a dataset by calculating summary statistics (mean,
median, mode, standard deviation) for numerical columns.


```python
# Statistics for numerical columns
summary_stats = df[['Quantity', 'UnitPrice', 'NetSales']].agg(['mean', 'median', 'std'])

# Mode needs separate call
summary_modes = df[['Quantity', 'UnitPrice', 'NetSales']].mode().iloc[0]
summary_stats.loc['mode'] = summary_modes

print("Summary Statistics:\n", summary_stats)
```

30. Perform feature engineering on a dataset by creating new features and then visualize their importance using a bar chart.

```python
#Feature Engineering
df['Discount_Amount'] = df['UnitPrice'] * df['Quantity'] * (df['DiscountPct'] / 100)
```

```python
#Feature Engineering is used to enhance the readablity of the data as here we can get exact how much discount amount is get from particular order.
```

```python
cat_discount = df.groupby('Category')['Discount_Amount'].sum().reset_index()

# Visualization
plt.figure(figsize=(8, 5))
sns.barplot(data=cat_discount, x='Category', y='Discount_Amount', palette='viridis')

plt.title('Total Discount Amount by Category', fontsize=14, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Discount Amount ($)', fontsize=12)
plt.tight_layout()
plt.show()
```
