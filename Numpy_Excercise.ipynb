29. Solve the system of linear equations 2x + 3y = 5 and 4x + 6y = 10 using NumPy.


```python
a29 = np.array([[2, 3], [4, 5]])
b29 = np.array([5, 10])
print("Coefficient matrix:\n LHS:", a29, " RHS:", b29)
try:
    solution29 = np.linalg.solve(a29, b29)
    print("Solution:", solution29)
except np.linalg.LinAlgError as e:
    print("Singular matrix error:", e)
    print("(The two equations are linearly dependent tends to infinite solutions.)")
    lstsq_solution, *_ = np.linalg.lstsq(a29, b29, rcond=None)
    print("Least-squares (minimum norm) solution:", lstsq_solution)
```

```python
print("Example for singular matrix\n")
a29_1 = np.array([[2, 3], [4, 6]])
b29_2 = np.array([5, 10])
print("Coefficient matrix:\n LHS:", a29_1, " RHS:", b29_2)
try:
    solution29_a = np.linalg.solve(a29_1, b29_2)
    print("Solution:", solution29_a)
except np.linalg.LinAlgError as e:
    print("Singular matrix error:", e)
    print("(The two equations are linearly dependent tends to infinite solutions.)")
    lstsq_solution, *_ = np.linalg.lstsq(a29_1, b29_2, rcond=None)
    print("Least-squares (minimum norm) solution:", lstsq_solution)
```

30. Perform Singular Value Decomposition (SVD) on a 3x3 matrix and reconstruct the original matrix using the SVD components.

```python
mat30 = np.array([[1, 2, 3], [4, 5, 6], [7, 8, 10]])
U30, S30, VT30 = np.linalg.svd(mat30)
Sigma30 = np.zeros((3, 3))
np.fill_diagonal(Sigma30, S30)
reconstructed30 = U30 @ Sigma30 @ VT30
print("Original matrix:\n", mat30)
print("U:\n", U30)
print("Singular values:", S30)
print("VT:\n", VT30)
print("Reconstructed matrix:\n", np.round(reconstructed30, 8))
```
